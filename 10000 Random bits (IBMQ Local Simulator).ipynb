```python
from qiskit import QuantumCircuit, execute, Aer
```

```python
# Are's quasm simulator
simulator = Aer.get_backend('qasm_simulator')
```

```python
# Create a quantum circuit and attach H gate to qbit 0
circuit = QuantumCircuit(1,1)
circuit.h(0)

#Measure the state of qbit
circuit.measure(0 , 0)

#Iterate loop 10000 times to build 10000 bit random string
rand_string = ""
for _ in range(10000):
    job = execute(circuit, simulator, shots=1)
    result = job.result()
    d = list(result.get_counts().keys())
    rand_string += d[0]
print(rand_string)
```

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute,IBMQ
from qiskit.tools.monitor import job_monitor
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider(group='open')
```

```python
q = QuantumRegister(5,'q')
c = ClassicalRegister(5,'c')
circuit = QuantumCircuit(q,c)
circuit.h(q) # Applies hadamard gate to all qubits
circuit.measure(q,c) # Measures all qubits 
```

```python
# This job is submitted to IBMQ ESSEX. It is a 5 bit quantum computer 
backend = provider.get_backend('ibmq_essex')
#shots = 2000 (2000*5 = 10000)
job1 = execute(circuit, backend, shots=2000)
job_monitor(job1)
result1 = job1.result()
rand_string1 = ""
rand_bits1 = result1.get_counts()
for each in rand_bits1:
    rand_string1 += each*rand_bits1[each]
print(rand_string1)
```

```python
# This job is submitted to IBMQ Burlington. It is a 5 bit quantum computer 
backend = provider.get_backend('ibmq_burlington')
#shots = 1000 (1000*20 = 10000)
job2 = execute(circuit, backend, shots=1000)
job_monitor(job2)
result2 = job2.result()
rand_string2 = ""
rand_bits2 = result2.get_counts()
for each in rand_bits2:
    rand_string2 += each*rand_bits2[each]
print(rand_string2)
```

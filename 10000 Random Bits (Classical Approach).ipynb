```python
import time
import itertools
import numpy
import matplotlib.pyplot as plt

seed = 0
def generate_random():
    randstring = ""
    global seed
    if seed == 0:
        seed = int(time.time())
    for _ in range(10000):
        rand_no = (seed*7+53)%2
        seed = ((seed+11)*7)%97
        randstring += str(rand_no)
    return randstring
if __name__ == "__main__":
    random_string = generate_random()
    print(random_string)
```
